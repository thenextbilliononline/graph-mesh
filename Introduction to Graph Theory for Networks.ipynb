# Introduction to Graph Theory for Networks

````python
import networkx as nx
import matplotlib.pyplot as plt
````

Graph Theory is the branch of mathematics that deals with things like maps, networks, flow, and centrality. Like most of mathematics, it sbtarts with a few basic building blocks that lead to some surprising results. 

In this tutorial, we will be using a Jupyter Notebook and networkx to explore graph theory. Other dependencies are listed in the file.

````python
node = nx.Graph()
node.add_node(1)
nx.draw(node,node_color='black')
````

This is a **node**. It is a basic building block of graph theory. It can be represented as a dot, point, or area on a page. A collection of one or more nodes is called a **graph**. While it seems abstract, nodes can be used to represent access points on a wireless network, species in an ecosytem, or stops along a supply chain. We will focus on the first such type of node.

````python
graph = nx.Graph()
this_list = range(10)
graph.add_nodes_from(this_list)
nx.draw(graph,node_color='black')
````

This is a **graph**. It is a collection of more than one **node**. In a network, two nodes can be connected as in the figure below. That connection is called a **vertex** or **edge**. This is an **unconnected** graph.

````python
nodes = nx.Graph()
nodes.add_node(1)
nodes.add_node(2)
nodes.add_edge(1,2)
nx.draw(nodes,node_color='black')
````

## Networks



````python
graph = nx.Graph()
count = range(10)
this_list = []
for i in range(10):
    for j in range(10):
        this_list.append([i,j])
graph.add_nodes_from(count)
graph.add_edges_from(this_list)
nx.draw(graph,node_color='black')
````

This is a **fully-connected network** or a **complete graph**. Networkx (nx) gives us a tool to generate these easily.

````python
G = nx.complete_graph(10)
nx.draw(G,node_color='black')
````

## Types of Nodes

**Maps** are a special type of network in which none of the edges of graph have to cross. These are the types of networks that are most interesting to us because they correspond to non-overlapping 'cells' surround by edges and vertices.

Hub
Branch
Leaf

````python
graph = nx.Graph()
count = range(10)
this_list = []
for i in range(10):
    this_list.append([i,i+1])
graph.add_nodes_from(count)
graph.add_edges_from(this_list)
nx.draw(graph,node_color=['g','y','y','y','y','y','y','y','y','y','g'])
````

This is a **path**-- a network with only a signle route from one end to the other. In this type of graph, each node has a **connectivity** of either 1 or 2. 

A **leaf** node has only connection to the network, like the green nodes in the above network. The yellow nodes are called **branch** nodes. In a mesh network, leaf nodes can only provide or use **network services** (like bandwith or applications), while branch nodes allow information to **transit**, or move across, the network. For the sake of convenience within this document, leaf nodes will always be green and branch nodes will always be yellow.


Traditional telecommunications infrastructure often looks like this, particularly between towns or under oceans. The nodes in this case would represent routers, switches, and repaters. 

````python
graph = nx.Graph()
count = range(10)
this_list = []
for i in range(10):
    this_list.append([i,i+1])
graph.add_nodes_from(count)
graph.add_edges_from(this_list)
#
nx.draw(graph,node_color=['g','y','y','y','y','y','y','y','y','y','g'])
````


The problem with path graphs is that they are not resilient to network outages. A single node removal can change the entire structure of the network.

````python
graph = nx.Graph()
count = range(10)
this_list = []
for i in range(10):
    this_list.append([i,i+1])
graph.add_nodes_from(count)
graph.add_edges_from(this_list)
graph.remove_node(4) 
#The above line removes node '4', creating two disconnected networks
nx.draw(graph,node_color=['g','y','y','g','g','y','y','y','y','g'])
````

Path graphs, while useful for covering large distances cheaply, are not resilient. In the example above, we removed the 5th node with the ```remove_node( ) ``` function. We passed the number 4 because Python starts counting at ```0```, meaning that the 5th node is enumerated as ```4```.

````python
G = nx.star_graph(3)
nx.draw(G,node_color=['y','g','g','g'])
````

In some cases, a single point (the yellow node) can act as a local hub. This is called a **star graph** and can be called as above. These types of networks have the same problems as path graphs.

````python
wheel = nx.wheel_graph(9)
nx.draw(wheel,node_color='y')
````

This is another type of special graph called a **wheel graph**, which is characterized by two things: a circular **cycle** part and a hub-like **star** component. These networks create multiple redundant routes

````python
graph = nx.random_powerlaw_tree(10)
nx.draw(graph, node_color=['r','y','y','y','y','y','g','g','g','g'])
````

Many telecommunications networks have a single point that provides access to the Internet, that is denoted by a red node in the above diagram. The often have to extend that network access through many repeaters, routers, wired, and wireless devices (the yellow node) before it reaches customers in their homes and offices (green) nodes. This network topology combines the worst parts of path graphs as well as the single-point network hubs that are prone to outages. 

````python
graph = nx.Graph()
count = range(10)
this_list = []
for i in range(10):
    for j in range(10):
        this_list.append([i,j])
graph.add_nodes_from(count)
graph.add_edges_from(this_list)
nx.draw(graph,node_color=['r','r','y','y','y','r','y','r','y','r'])
````

This a **mesh network**. In this case, we see 10 nodes that are all able to   all connect with each other. Likewise, there are no **leaf** nodes and everyone can given as well as receive resources to or from the network. This type of network allows for any person to host their own network service, whether that's an email server, peer to peer Netflix, or access to the Internet.
